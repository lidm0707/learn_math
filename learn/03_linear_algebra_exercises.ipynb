### Exercise 13: Singular matrix (not invertible)

Try to invert `C = [[1,2],[2,4]]`. What happens? Why?

First check the determinant. If det=0, the matrix is **singular** (not invertible).

**Hint:** Use try/except with `np.linalg.inv()`

````python
# YOUR CODE HERE


# Expected output:
# det = 0.0
# Matrix is singular! Cannot invert.
````

### Solution

```python
C = np.array([[1,2],[2,4]])
det = np.linalg.det(C)
print(f"det = {det}")

if abs(det) < 1e-10:
    print("Matrix is singular! Cannot invert.")
    print("Row 2 = 2 * Row 1, so information was lost.")
else:
    print(np.linalg.inv(C))
```

### Exercise 16: Verify eigenvector equation

Using the same matrix `A = [[2,1],[1,2]]`, verify that `A @ v = lambda * v` for each eigenpair.

**Hint:** `eigenvectors[:, i]` gives the i-th eigenvector (column)

````python
# YOUR CODE HERE


# Expected output:
# Eigenvalue 3.0: A@v == lambda*v? True
# Eigenvalue 1.0: A@v == lambda*v? True
````

### Solution

```python
A = np.array([[2,1],[1,2]])
eigenvalues, eigenvectors = np.linalg.eig(A)

for i in range(len(eigenvalues)):
    v = eigenvectors[:, i]
    lam = eigenvalues[i]
    Av = A @ v
    lv = lam * v
    print(f"Eigenvalue {lam:.1f}: A@v == lambda*v? {np.allclose(Av, lv)}")
```

### Exercise 20: SVD and reconstruction

1. Create a 5x5 random matrix: `np.random.seed(42); A = np.random.randn(5, 5)`
2. Compute SVD: `U, S, Vt = np.linalg.svd(A)`
3. Reconstruct: `A_reconstructed = U @ np.diag(S) @ Vt`
4. Verify: `np.allclose(A, A_reconstructed)` should be `True`

**This proves SVD is a perfect decomposition - no information lost!**

````python
# YOUR CODE HERE


# Expected output:
# Original A:
# [[ 0.4967 ... ]]
# Singular values: [3.53 ... ]
# Reconstruction matches: True
````

### Solution

```python
np.random.seed(42)
A = np.random.randn(5, 5)

U, S, Vt = np.linalg.svd(A)

print(f"Original A:\n{A.round(4)}")
print(f"\nSingular values: {S.round(4)}")

A_reconstructed = U @ np.diag(S) @ Vt
print(f"\nReconstruction matches: {np.allclose(A, A_reconstructed)}")
```
