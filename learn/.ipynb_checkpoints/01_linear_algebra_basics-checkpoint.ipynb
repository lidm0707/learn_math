# Linear Algebra - From Zero to Understanding

## What is Linear Algebra?

Linear Algebra is the math of **vectors** and **linear transformations** (matrices).

Think of it as:
- **Vectors** = arrows with direction and length (data points)
- **Matrices** = machines that transform vectors (rotate, scale, shear)

### The Big Picture

| Concept | Think of it as | Used in |
|---------|---------------|----------|
| Vector | A point or arrow in space | Data representation |
| Matrix | A transformation machine | Transforming data |
| Dot Product | How much two arrows point same way | Similarity, projection |
| Determinant | How much area/volume changes | Invertibility check |
| Eigenvalues/Eigenvectors | Directions that don't rotate | PCA, compression |

````python
import numpy as np
````

---
## 1. Scalars, Vectors, and Spaces

### Scalar = Single number
```python
x = 5  # this is a scalar, just one number
```

### Vector = List of numbers = A point in space

A vector `[3, 4]` means:
- Go 3 units right on x-axis
- Go 4 units up on y-axis

It's a coordinate in space. That's it!

````python
# A vector is just an array of numbers
v = np.array([3, 4])
print(f"Vector v = {v}")
print(f"Dimension (how many numbers) = {v.ndim}")
print(f"Shape = {v.shape}")

# 3D vector
v3d = np.array([1, 2, 3])
print(f"\n3D vector = {v3d}")
print(f"This is a point in 3D space: x=1, y=2, z=3")
````

---
## 2. Vector Operations

### Addition = Move, then move again

If you walk 3 steps right + walk 1 step right = 4 steps right.
Vector addition works the same way component by component.

````python
a = np.array([1, 2])
b = np.array([3, 1])

# Addition: add each component
print(f"a = {a}")
print(f"b = {b}")
print(f"a + b = {a + b}")  # [4, 3]
print()
print("Think: start at origin, follow arrow a, then follow arrow b")
print("  a: go right 1, up 2")
print("  b: go right 3, up 1")
print("  total: right 4, up 3 => [4, 3]")
````

### Scalar Multiplication = Stretch or shrink

Multiply every component by the same number.

````python
v = np.array([2, 3])

print(f"v = {v}")
print(f"2 * v = {2 * v}")   # stretched to double
print(f"0.5 * v = {0.5 * v}") # shrunk to half
print(f"-1 * v = {-1 * v}")  # flipped direction
print()
print("Think of it as stretching or flipping the arrow")
````

---
## 3. Dot Product (The Most Important Operation)

### What it measures: **How much two vectors point in the same direction**

Formula: `a . b = a[0]*b[0] + a[1]*b[1] + ...`

### Intuition:
- **Positive** = vectors point roughly same direction (angle < 90)
- **Zero** = vectors are perpendicular (angle = 90)
- **Negative** = vectors point opposite directions (angle > 90)

### Why it matters:
- Machine Learning: cosine similarity, attention mechanisms
- Physics: work = force . distance
- Graphics: lighting calculations

````python
a = np.array([1, 0])  # pointing right
b = np.array([0, 1])  # pointing up
c = np.array([1, 1])  # pointing diagonal
d = np.array([-1, 0]) # pointing left

print("Dot product examples:")
print(f"  right . up     = {np.dot(a, b):3.0f}  => perpendicular (90 degrees)")
print(f"  right . diag   = {np.dot(a, c):3.0f}  => same-ish direction")
print(f"  right . left   = {np.dot(a, d):3.0f}  => opposite direction")
print(f"  right . right  = {np.dot(a, a):3.0f}  => same direction")
print()
print("Rule of thumb:")
print("  dot > 0 => acute angle (friends)")
print("  dot = 0 => right angle (strangers)")
print("  dot < 0 => obtuse angle (enemies)")
````

---
## 4. Vector Length (Norm)

### Formula: `|v| = sqrt(v . v)` = `sqrt(v[0]^2 + v[1]^2 + ...)`

This is just **Pythagoras' theorem** extended to any dimension!

For `[3, 4]`: length = sqrt(9 + 16) = sqrt(25) = **5**

````python
v = np.array([3, 4])

length = np.linalg.norm(v)
print(f"Vector v = {v}")
print(f"Length |v| = sqrt(3^2 + 4^2) = sqrt(9+16) = sqrt(25) = {length}")

# Unit vector (length = 1) - very important concept!
unit_v = v / length
print(f"\nUnit vector = {unit_v}")
print(f"Unit vector length = {np.linalg.norm(unit_v):.4f}")
print()
print("Unit vector = direction without magnitude")
print("Think: 'which way' without 'how far'")
````

---
## 5. Matrices as Transformations

### The KEY insight: A matrix is a FUNCTION that transforms vectors

When you multiply `matrix @ vector`, you're applying a transformation.

### The columns of a matrix tell you WHERE the basis vectors go:

```
Matrix = [[a, c],    Column 1 = [a, b] = where [1,0] goes
          [b, d]]    Column 2 = [c, d] = where [0,1] goes
```

### Common transformations:
- **Identity** = change nothing
- **Scale** = stretch/shrink axes
- **Rotation** = rotate by angle
- **Shear** = push sideways
- **Reflection** = mirror

````python
v = np.array([1, 2])
print(f"Original vector: {v}")

# Identity - does nothing
I = np.array([[1, 0],
              [0, 1]])
print(f"\nAfter Identity:  {I @ v}  (same!)")

# Scale x2 in x, x3 in y
S = np.array([[2, 0],
              [0, 3]])
print(f"After Scale:    {S @ v}  (x*2, y*3)")

# Rotation 90 degrees counter-clockwise
R90 = np.array([[ 0, -1],
                [ 1,  0]])
print(f"After Rot 90:   {R90 @ v}  (rotated!)")

# Shear (push x based on y)
H = np.array([[1, 1],
              [0, 1]])
print(f"After Shear:    {H @ v}  (x += y)")

# Reflection across y-axis
F = np.array([[-1, 0],
              [ 0, 1]])
print(f"After Reflect:  {F @ v}  (flip x)")
````

---
## 6. Matrix Multiplication = Composition of Transformations

### When you multiply two matrices, you're COMBINING transformations.

If A = rotate 90, and B = scale by 2, then:
- `A @ B` = first scale, then rotate
- `B @ A` = first rotate, then scale

**Order matters!** `A @ B` != `B @ A` in general.

````python
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[2, 0],
              [0, 3]])

print(f"A =\n{A}")
print(f"\nB =\n{B}")
print(f"\nA @ B =\n{A @ B}")
print(f"\nB @ A =\n{B @ A}")
print(f"\nAre they equal? {np.array_equal(A @ B, B @ A)}")
print()
print("Matrix multiplication is NOT commutative!")
print("Order of transformations matters.")
````

---
## 7. Determinant = How Much Area Changes

### For 2x2 matrix [[a,b],[c,d]]: det = ad - bc

### What it tells you:
- **det > 0**: Transformation preserves orientation (no flipping)
- **det < 0**: Transformation flips orientation (mirror)
- **det = 0**: Transformation **collapses space** into lower dimension

### The last one is HUGE:
If det = 0, you LOSE information. The transformation is not reversible.
Think of it like crushing a 3D box into a flat 2D paper - you can't get back.

````python
# Scale: area multiplied by 6 (2 * 3)
S = np.array([[2, 0],
              [0, 3]])
print(f"Scale matrix:\n{S}")
print(f"Determinant = {np.linalg.det(S):.0f}")
print(f"Area changes by factor of {np.linalg.det(S):.0f}\n")

# Rotation: area preserved
theta = np.pi / 4  # 45 degrees
R = np.array([[np.cos(theta), -np.sin(theta)],
              [np.sin(theta),  np.cos(theta)]])
print(f"Rotation 45 matrix:\n{R.round(4)}")
print(f"Determinant = {np.linalg.det(R):.4f}")
print(f"Rotation preserves area!\n")

# Singular (collapses space)
C = np.array([[1, 2],
              [2, 4]])  # row2 = 2 * row1
print(f"Singular matrix:\n{C}")
print(f"Determinant = {np.linalg.det(C):.4f}")
print(f"det=0 means: space collapsed! NOT invertible!")
````

---
## 8. Matrix Inverse = Undo a Transformation

### If A transforms v, then A^(-1) transforms it back.

```
A @ v = w
A^(-1) @ w = v   (undo!)
```

### Requirements:
- det(A) != 0 (you can't undo a collapse)
- Only square matrices can be inverted

````python
A = np.array([[1, 2],
              [3, 4]])

print(f"Original matrix A:\n{A}")
print(f"det(A) = {np.linalg.det(A):.1f}\n")

A_inv = np.linalg.inv(A)
print(f"Inverse A^(-1):\n{A_inv.round(4)}\n")

# Proof: A @ A_inv = Identity
product = A @ A_inv
print(f"A @ A^(-1) = \n{product.round(10)}")
print("^ This is the Identity matrix! Transformation undone.\n")

# Practical example
v = np.array([5, 6])
w = A @ v  # transform
v_back = A_inv @ w  # undo
print(f"v = {v}")
print(f"After A: w = {w}")
print(f"After A^(-1): {v_back} (back to original!)")
````

---
## 9. Solving Systems of Linear Equations

### This is THE application of linear algebra.

Given:
```
2x + 3y = 8
 x + 2y = 5
```

Rewrite as matrix form:
```
[[2, 3],  @ [x,   = [8,
 [1, 2]]    y]      5]
```

So: **A @ x = b**, solve for **x**

Solution: **x = A^(-1) @ b**

````python
# System: 2x + 3y = 8, x + 2y = 5
A = np.array([[2, 3],
              [1, 2]])
b = np.array([8, 5])

# Method 1: Using inverse
x = np.linalg.inv(A) @ b
print(f"Solution: x={x[0]:.1f}, y={x[1]:.1f}")

# Verify
print(f"\nVerify: 2({x[0]:.1f}) + 3({x[1]:.1f}) = {2*x[0] + 3*x[1]:.1f}")
print(f"Verify:  ({x[0]:.1f}) + 2({x[1]:.1f}) = {x[0] + 2*x[1]:.1f}")

# Method 2: Using np.linalg.solve (BETTER - more numerically stable)
x2 = np.linalg.solve(A, b)
print(f"\nUsing solve: x={x2[0]:.1f}, y={x2[1]:.1f}")
print("(Same answer, but solve() is more stable for large systems)")
````

## 10. Eigenvalues & Eigenvectors

### The most abstract but powerful concept.

### Question: When we apply transformation A, are there vectors that only STRETCH (not rotate)?

### Answer: Yes! Those are **eigenvectors**, and the stretch factor is the **eigenvalue**.

```
A @ v = lambda * v
  ^       ^    ^
matrix  same   eigenvalue (just a number)
         dir
         eigenvector
```

### Why it matters:
- **PCA** (dimensionality reduction) = find principal eigenvectors
- **Google PageRank** = find dominant eigenvector
- **Vibration analysis** = eigenvalues = natural frequencies
- **Quantum mechanics** = eigenvalues = observable quantities

````python
A = np.array([[4, 2],
              [1, 3]])

eigenvalues, eigenvectors = np.linalg.eig(A)

print(f"Matrix A:\n{A}\n")
print(f"Eigenvalues: {eigenvalues}")
print(f"Eigenvectors (columns):\n{eigenvectors.round(4)}\n")

# Verify: A @ v = lambda * v
for i in range(len(eigenvalues)):
    v = eigenvectors[:, i]
    lam = eigenvalues[i]
    Av = A @ v
    lv = lam * v
    print(f"Eigenvalue {lam:.2f}:")
    print(f"  A @ v = {Av.round(4)}")
    print(f"  lam*v = {lv.round(4)}")
    print(f"  Equal? {np.allclose(Av, lv)}")
    print()
````

## 11. Span, Linear Independence, Basis, Rank

### Span = All points you can reach with linear combinations

`a*v1 + b*v2` where a,b are any real numbers.

### Linear Independence
Vectors are independent if NONE of them can be written as a combination of others.

Example:
- `[1,0]` and `[0,1]` -> independent (need both to reach any point)
- `[1,0]` and `[2,0]` -> dependent (second is just 2x first)

### Basis
A set of independent vectors that span the whole space.
Standard basis for 2D: `[1,0]` and `[0,1]`

### Rank
Number of independent rows/columns in a matrix.
Rank = dimension of the output space.

````python
# Rank examples
print("=== Rank Examples ===\n")

# Full rank (rank = 2 for 2x2 matrix)
A = np.array([[1, 2],
              [3, 4]])
print(f"A =\n{A}")
print(f"Rank = {np.linalg.matrix_rank(A)}")
print(f"det = {np.linalg.det(A):.1f} (non-zero = full rank)\n")

# Not full rank (rows are dependent: row2 = 2*row1)
B = np.array([[1, 2],
              [2, 4]])
print(f"B =\n{B}")
print(f"Rank = {np.linalg.matrix_rank(B)}")
print(f"det = {np.linalg.det(B):.1f} (zero = NOT full rank)")
print(f"Row 2 is 2x Row 1, so we only have 1 independent row")
````

## 12. Real-World Example: Image as Matrix

An image is just a matrix of pixel values!
- Grayscale image: 2D matrix (height x width)
- Color image: 3D tensor (height x width x 3 for RGB)

This is why linear algebra is fundamental to image processing, computer vision, and deep learning.

````python
# Create a simple "image" as a matrix
image = np.array([
    [  0,  50, 100, 150, 200, 255],
    [ 50, 100, 150, 200, 255, 200],
    [100, 150, 200, 255, 200, 150],
    [150, 200, 255, 200, 150, 100],
    [200, 255, 200, 150, 100,  50],
    [255, 200, 150, 100,  50,   0],
])

print("Grayscale image as matrix (6x6 pixels):")
print(image)
print(f"\nShape: {image.shape}")
print(f"Min pixel: {image.min()}, Max pixel: {image.max()}")
print()
print("Every photo on your phone is a HUGE matrix of numbers.")
print("Image filters = matrix operations!")
print("Neural networks process images as matrix multiplications!")
````

## 13. Real-World Example: Recommendation System

Users rate movies 1-5. Each user is a **vector** of ratings.

```
         Movie1  Movie2  Movie3
Alice      5       3       1
Bob        1       2       5
Carol      4       3       2
```

**Dot product** of two user vectors = how similar their taste is.
Higher dot product = more similar taste.

````python
# User-Movie rating matrix
ratings = np.array([
    [5, 3, 1],  # Alice
    [1, 2, 5],  # Bob
    [4, 3, 2],  # Carol
])

users = ['Alice', 'Bob', 'Carol']

print("Rating Matrix:")
print("      Movie1 Movie2 Movie3")
for i, name in enumerate(users):
    print(f"{name:6} {ratings[i][0]:5} {ratings[i][1]:5} {ratings[i][2]:5}")

print("\n=== Similarity (Dot Product) ===")
for i in range(3):
    for j in range(i+1, 3):
        sim = np.dot(ratings[i], ratings[j])
        print(f"  {users[i]} . {users[j]} = {sim:3d}")

print("\nConclusion: Alice and Carol are most similar.")
print("If Alice liked a new movie, recommend it to Carol!")
````

## 14. SVD (Singular Value Decomposition) - The Swiss Army Knife

### Any matrix A can be broken into 3 pieces:

```
A = U @ S @ V^T
```

- **U** = where input vectors land (rotation)
- **S** = how much to scale (diagonal, singular values)
- **V^T** = how to rotate input

### Why it matters:
- **Image compression**: keep only big singular values
- **PCA**: find most important directions in data
- **Netflix Prize**: was won using SVD on user-movie matrix
- **LSA** (search engines): find latent topics in documents

````python
A = np.array([[3, 2, 2],
              [2, 3, -2]])

print(f"Original matrix A ({A.shape}):\n{A}\n")

U, S, Vt = np.linalg.svd(A, full_matrices=False)

print(f"U ({U.shape}) - left singular vectors:\n{U.round(4)}\n")
print(f"Singular values: {S.round(4)}")
print(f"(Bigger = more important direction)\n")
print(f"Vt ({Vt.shape}) - right singular vectors:\n{Vt.round(4)}\n")

# Reconstruct to verify
A_reconstructed = U @ np.diag(S) @ Vt
print(f"Reconstructed A:\n{A_reconstructed.round(4)}")
print(f"Matches original? {np.allclose(A, A_reconstructed)}")
````

## 15. Summary Cheat Sheet

| Operation | Code | Meaning |
|-----------|------|---------|
| Vector add | `a + b` | Combine movements |
| Scalar mul | `2 * v` | Stretch/shrink |
| Dot product | `np.dot(a, b)` | Similarity measure |
| Vector length | `np.linalg.norm(v)` | Size of vector |
| Matrix mul | `A @ B` | Compose transforms |
| Transform | `A @ v` | Apply transformation |
| Determinant | `np.linalg.det(A)` | Area change factor |
| Inverse | `np.linalg.inv(A)` | Undo transformation |
| Solve system | `np.linalg.solve(A, b)` | Find x where Ax=b |
| Eigen | `np.linalg.eig(A)` | Unchanging directions |
| SVD | `np.linalg.svd(A)` | Factor any matrix |
| Rank | `np.linalg.matrix_rank(A)` | Independent dimensions |

### The Learning Path:
1. **Vectors** -> points in space
2. **Dot product** -> similarity / projection
3. **Matrices** -> transformations
4. **Determinant** -> area change / invertibility
5. **Inverse** -> undo transformations
6. **Eigen** -> natural directions of transformation
7. **SVD** -> decompose anything
